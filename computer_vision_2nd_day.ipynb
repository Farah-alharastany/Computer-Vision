```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
imge = cv2.imread("PYTHON.jpg")
gray = cv2.cvtColor(imge, cv2.COLOR_BGR2GRAY)
plt.imshow(imge),plt.show()
corner = cv2.goodFeaturesToTrack(gray, 30, 0.01,10)
corner = np.int0(corner)
for i in corner:
  x,y =i.ravel()
  cv2.circle(imge, (x,y), 3,255, -1)
plt.imshow(imge)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
imge = cv2.imread("download.png")
gray = cv2.cvtColor(imge, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
des = cv2.cornerHarris(gray, 2, 5, 0.07)
des = cv2.dilate(des, None) 
#change the color of corners to red, the corner is grater that 0.01
imge[des >0.01 * des.max()] = [255, 0, 0]
plt.imshow(imge)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import requests
from PIL import Image
#get the image
imge = Image.open(requests.get('https://a57.foxnews.com/media.foxbusiness.com/BrightCove/854081161001/201805/2879/931/524/854081161001_5782482890001_5782477388001-vs.jpg', stream = True). raw)
#resize the image
imge = imge.resize((450, 250))
#conert the image into an array
img_arr =np.array(imge)
#convert the image into gray 
gray = cv2.cvtColor(img_arr, cv2.COLOR_BGR2GRAY)
#to make the image better in processing تنعيم الصورة
blur = cv2.GaussianBlur(gray, (5,5), 0)
Image.fromarray(blur)
#to wider the image, توسيع الصورة
dilated = cv2.dilate(blur, np.ones((3,3)))
Image.fromarray(dilated)
plt.imshow(gray)
#لملء عناصر البكسل بالعنصر, لمنع التاكل بالصورة
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2,2))
closing = cv2.morphologyEx(dilated, cv2.MORPH_CLOSE,kernel)
Image.fromarray(closing)
#trained model
Car_SRC ='haarcascade_car.xml'
car = cv2.CascadeClassifier(Car_SRC)
cars = car.detectMultiScale(closing, 1.1,1 )
cnt =0
for(x,y,w,h) in cars:
  #(255,0,0) the color, 2 is the thickness of the rectangle
  cv2.rectangle(img_arr, (x,y), (x+w,y+h), (255,0,0), 2)
  cnt+=1
  print(cnt, "cars")
  Image.fromarray(img_arr)
plt.imshow(img_arr)
```
